2.  How does broadcasting work in NumPy?

 In NumPy, broadcasting is a technique that allows arrays of different shapes to be used together in arithmetic operations without explicitly reshaping them. Instead of repeating data to match shapes, NumPy automatically expands smaller arrays along missing dimensions.

How it works:

Alignment of Shapes: NumPy compares the shapes of the arrays from the rightmost dimension to the left.

Dimension Matching:

If two dimensions are equal, they are compatible.

If one of the dimensions is 1, it gets "broadcasted" (expanded) to match the other dimension.

If the dimensions are different and neither is 1, broadcasting fails.

Example:

      import numpy as np

        A = np.array([[1, 2, 3]])  # Shape (1, 3)
        B = np.array([[4], [5], [6]])  # Shape (3, 1)

        result = A + B  # Shape (3, 3) after broadcasting
        print(result)
      Output:

          [[5 6 7]
          [6 7 8]
          [7 8 9]]

Here, A is expanded to (3,3) and B is also expanded to (3,3), enabling element-wise addition.

Advantages:

Memory efficiency: No unnecessary data replication.

Faster computations: Optimized operations with minimal overhead.

8.  What does the term “vectorized operation” mean in NumPy?

 In NumPy, a vectorized operation refers to the ability to perform operations on entire arrays (or vectors) at once, rather than using loops to process individual elements. This is possible because NumPy internally uses optimized C code that can execute these operations much faster than Python’s standard loop-based methods.

For example, if you want to add two arrays element-wise, you can do it directly with a vectorized operation like this:

      import numpy as np
      arr1 = np.array([1, 2, 3])
      arr2 = np.array([4, 5, 6])

     result = arr1 + arr2  # vectorized operation

In this case, arr1 + arr2 is a vectorized operation, and NumPy will automatically apply the addition to each corresponding element of the arrays, rather than looping through them manually.

Vectorized operations are highly efficient, both in terms of speed and memory usage, because they avoid the overhead of explicit loops and take advantage of low-level optimizations in NumPy.

21.  Why is NumPy’s array slicing faster than Python’s list slicing?

 NumPy's array slicing is faster than Python’s list slicing because:

Memory Efficiency: NumPy arrays use a contiguous block of memory, making element access and slicing much faster than Python lists, which store references to objects.

View Instead of Copy: When slicing a NumPy array, it creates a view of the original data instead of copying it, reducing memory usage and improving performance. In contrast, slicing a Python list creates a new list with copies of the elements.

Optimized C Implementation: NumPy is implemented in C and optimized for numerical computations, enabling much faster operations than Python’s built-in list operations.

Because of these optimizations, NumPy is widely used for data analysis and scientific computing where speed is crucial.

```python
#4.  How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample data
data = {
    'A': [1, 2, 3, 4, 5],
    'B': [2, 3, 4, 5, 6],
    'C': [5, 4, 3, 2, 1],
    'D': [1, 3, 5, 7, 9]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Compute the correlation matrix
corr_matrix = df.corr()

# Create a heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix Heatmap")
plt.show()
```
